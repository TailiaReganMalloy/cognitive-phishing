```python
import os
from llm_email_gen.ollama_cognitive_phishing_rag import CognitivePhishingRAG
```

```python
dataset_path = "data/Emails_Cognitive.csv"
definitions_path = "cognitive_bias_definitions.json"
generated_email = ""
if not os.path.exists(dataset_path) or not os.path.exists(definitions_path):
    print(f"Error: Ensure '{dataset_path}' and '{definitions_path}' exist in your working directory.")
else:
    rag_system = CognitivePhishingRAG(dataset_path, definitions_path)
    sample_base_prompt = "Draft an email to an Adobe Creative cloud user with the e-mail id: joeydiaz@gmail.com."
    sample_biases = ["Anchoring effect"]
    print(f"Generating email using biases: {sample_biases}...\n")
    try:
        generated_email = rag_system.generate_email(
            base_prompt=sample_base_prompt, 
            target_biases=sample_biases,
            top_k=2,
            model="gemma3:27b"
        )
        print("=== GENERATED SYNTHETIC PHISHING EMAIL ===")
        print(generated_email)
    except Exception as e:
        print(f"An error occurred during generation: {e}")
```

```python
dataset_path = "data/Emails_Cognitive.csv"
definitions_path = "cognitive_bias_definitions.json"
base_prompt = "An email from IT support asking the user to update their VPN client by clicking a link, or they will lose remote access."
target_biases = ["Authority bias", "Urgency effect", "Loss aversion"]
print("-----------------------\n")

evaluator = CognitivePhishingEvaluator(definitions_path)

print("Evaluating email effectiveness...")
evaluation = evaluator.evaluate_email(generated_email, target_biases, model="gemma3:27b")

if evaluation:
    print("\n--- Evaluation Results ---")
    print(f"Overall Effectiveness: {evaluation.get('Overall_Effectiveness')}/10")
    print(f"Authenticity: {evaluation.get('Authenticity_Score')}/10 - {evaluation.get('Authenticity_Reasoning')}")
    print(f"Bias Integration: {evaluation.get('Bias_Integration_Score')}/10 - {evaluation.get('Bias_Integration_Reasoning')}")
    print(f"Evasion Score: {evaluation.get('Evasion_Score')}/10 - {evaluation.get('Evasion_Reasoning')}")
    print(f"Overall Reasoning: {evaluation.get('Overall_Reasoning')}")
```
